```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, MultiLabelBinarizer
```

```python
evidences = pd.read_json('release_evidences.json').T

evidences
```

```python
evidences = evidences[['name', 'possible-values', 'data_type']]

evidences
```

```python
print(type(evidences))
```

```python
feature_set = []

for evidence in evidences.itertuples():
    e_id, code, values, d_type = evidence 
    if d_type == 'B':
        feature_set.append(code)
    else:
        for value in values:
            feature_set.append(f'{code}_@_{value}')
feature_set
```

```python
len(feature_set)
```

```python
feature_set = pd.api.types.CategoricalDtype(
  categories=feature_set,
  ordered=True
)
# feature_set = np.expand_dims(feature_set, axis=1)

feature_set
```

```python
feature_map = dict(enumerate(pd.Series(['']).astype(feature_set).cat.categories))
feature_map
```

```python
# Encode feature
trans = np.expand_dims(feature_set, axis=1)
print(trans.shape, trans.ndim)
evidenceEncoder = OneHotEncoder()

evidenceEncoder.fit(trans)

evidenceEncoder
```

```python
mBEncoder = MultiLabelBinarizer()

mBEncoder.fit(trans)

mBEncoder
```
